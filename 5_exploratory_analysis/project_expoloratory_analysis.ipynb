## Предобработка данных

В полученных данных замечены пропуски, можем заполнить для следующих столбцов: `balcony`, `is_apartment`, `living_area`, `kitchen_area`.

В данных треуется изменить тип данных для стобца `balcony` так как количество балконов не может быть числом с плавающей точкой, также заменим тип данных для столбца `last_price` поскольку цены за недвижимость как правило без коппек выставляются на продажу и как следствие должен быть тип данных `int`, дополнительно можем поменять тип данных в столбце `first_day_exposition` на привычный `datetime64`

В остальных столбцах данные не заполнить, возможно потребуется отбросить некоторые строки на следующих этапах.

```python
# заполнение отсутствующих данных в столбце `balcony` цифрой 0 
data['balcony'] = data['balcony'].fillna(0)
# заполнение пропусков в столбце `is_apartment`
data['is_apartment'] = data['is_apartment'].fillna(False)
# заполнение пропусков в столбце`living_area`, посчитаем средний коэффициент и перемножим на значения `total_area`
koef_living_area = round(data['living_area'].mean() / data['total_area'].mean(), 2)
data['living_area'] = data['living_area'].fillna(data['total_area'] * koef_living_area)
# заполнение пропусков в столбце`kitchen_area`, посчитаем средний коэффициент и перемножим на значения `total_area`
koef_kitchen_area = round(data['kitchen_area'].mean() / data['total_area'].mean(), 2)
data['kitchen_area'] = data['kitchen_area'].fillna(data['total_area'] * koef_kitchen_area)
# расчет медианного значения для столбца `ceiling_height`
ceiling_height_median = data['ceiling_height'].median()
# заполнение отсутствующих данных в столбце `ceiling_height`
data['ceiling_height'] = data['ceiling_height'].fillna(ceiling_height_median)
```

```python
# изменение типа данных в столбце `balcony`
data['balcony'] = data['balcony'].astype('int')
# изменение типа данных в столбце `last_price`
data['last_price'] = data['last_price'].astype('int')
# изменение типа данных в столбце `first_day_exposition`
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'])
# получение обновленной общей информации о данных
display(data.info())
# получение первых пяти строк
data.head()
```

### Вывод

Пропуски заполнены там где это было возможно, а также заменены типы данных на привычные и характерные. Возможно некорректные типы данных были присвоены изза технической ошибки, а пропуски в строках указывают на то, что не все люди тратят врямя на подробную инфорацию в объявлении о продаже.

## Расчёты и добавление результатов в таблицу

Внесем дополнительные значения в таблицу о соотношении жилой площади к общей, площади кухни к общей, также потребуется столбец с ценой на квадратный метр, информация с годом, месяцем и днем публикации объявления. И укрупним группы по названию этажей: первый, последний или другой в зависимости от этажа размещения квартиры и количества этажей во всем доме.

```python
# получение столбца с ценой за квадратный метр
data['square_price'] = data['last_price'] / data['total_area']
# получение столбца с соотношением жилой площади к общей
data['div_living_area'] = data['living_area'] / data['total_area']
# получение столбца с соотношением площади кухни к общей
data['div_kitchen_area'] = data['kitchen_area'] / data['total_area']
# получение столбца с годом публикации
data['year_public'] = data['first_day_exposition'].dt.year
# получение столбца с месяцем публикации
data['month_public'] = data['first_day_exposition'].dt.month
# получение столбца с днем публикации
data['day_public'] = data['first_day_exposition'].dt.weekday
# получение столбца с указанием этажа через написание функции и ее применения к каждой строке `data`
def floor_name(row):
    floor = row['floor']
    floors_total = row['floors_total']
    if floor == 1:     
        return 'Первый'
    if floor == floors_total and floors_total != 1: 
        return 'Последний'
    else:
        return 'Другой' 
data['floor_name'] = data.apply(floor_name, axis=1)
# получение первых пятнадцати строк
data.head(15)
```

### Вывод

Требующиеся данные для дальнейшего анализа внесены.

### Удаление редких и выбивающихся значений

Для выявления редких и выбивающихся значений обратимся к пункту 4.1 и сохраним значения, удовлетворяющие условиям:
- Общая площадь от 12м2 и не более 200м2
- Цена не менее 200 000 руб. и не более 100 000 000 руб.
- Количество комнат до 5 включительно
- Высота потолков не менее 1,5м и до 6м

```python
# создание таблицы с отфильтрованными данными
filtered_data = data.query('12 <= total_area < 200 and 200000 <= last_price < 100000000 and rooms <= 5 and 1.5 <= ceiling_height < 6')
# расчет соотношения отфильтрованный данных к исходным
display(filtered_data['total_area'].count() / data['total_area'].count())
# получение первых десяти строк
filtered_data.info()
```

Выбросы составляли 2% от исходных данных, что не превышает 10% 
1. Есть квартиры с крайне большой площадью, напрмер максимальное значение 900м2 это крайне редкое значение. 
2. Минимально обнаруженная цена даже для области 12 000 руб., что очень странно даже для объявлений в 2014 году.

### Изучение 10 населённых пунктов с наибольшим числом объявлений

- Выберем 10 населённых пунктов с наибольшим числом объявлений и посчитаем среднюю цену квадратного метра в этих населённых пунктах. 
- Определим населённые пункты с самой высокой и низкой стоимостью жилья.

```python
# получение сводной таблицы с расчетом средней стоимости за квадратный метр и наибольшего количества объявлений
filtered_data_top_count = filtered_data.pivot_table(index='locality_name', values='square_price', aggfunc=['count', 'mean'])
# изменение названий столбцов
filtered_data_top_count.columns = ['count', 'square_price_average']
# сортировка значений по количеству объявлений от большего к меньшему и получение первых десяти строк
filtered_data_top_count = filtered_data_top_count.sort_values(by='count', ascending=False).head(10)
# вывод на экран полученной таблицы
display(filtered_data_top_count)
# сортировка значений по средней цене за квадратный метр от большего к меньшему и получение первых десяти строк
display(filtered_data_top_count.sort_values(by='square_price_average', ascending=False))
# визуализация таблицы с наибольшим количеством объявлений
filtered_data_top_count.plot(y='square_price_average', kind='bar')
```

### Расчет области входящей в центр Санкт-Петербурга

- Изучим предложения квартир: для каждой квартиры есть информация о расстоянии до центра и выделим квартиры в Санкт-Петербурге. 
- Выясним, какая область входит в центр с созданием столбца с расстоянием до центра в километрах: округлим до целых значений. 
- Посчитаем среднюю цену для каждого километра. 
- Построем график: показывающий, как цена зависит от удалённости от центра. 
- Определим границу, где график сильно меняется — это и будет центральная зона.

```python
# создание таблицы только с данными по Санкт-Петербургу
filtered_data_spb = filtered_data.query('locality_name == "Санкт-Петербург"').dropna(subset=['cityCenters_nearest'])
# создание столбца с расстоянием от центра в километрах, округлим до целого
filtered_data_spb['cityCenters_nearest_km'] = round(filtered_data_spb['cityCenters_nearest'] / 1000, 0)
# создание столбца с расчетом стоимости цены от километра
filtered_data_spb['cityCenters_nearest_km_price'] = filtered_data_spb['last_price'] / filtered_data_spb['cityCenters_nearest_km']
# создание сводной таблицы по медианным значениям по каждому км
filtered_data_spb_pivot = filtered_data_spb.pivot_table(index='cityCenters_nearest_km', values='square_price', aggfunc='median')
# построим график по медианным значениям по каждому км
filtered_data_spb_pivot.plot(kind='bar')
```

#### Вывод

По полученному графику видно, что резкий перепад есть между 6 и 7 километром, а также между 20 и 21 километром. В данном случае центром города можно считать границу в 6 км.
